```python
class Student(object):
    
    def __init__(self,last_name,first_name,id_number,major):
        self.last_name = str(last_name)
        self.first_name = str(first_name)
        self.id_number = str(id_number)
        self.major = str(major)
    
    def __repr__(self):
        return self.last_name + ", " + self.first_name
```

```python
s1 = Student("Smith", "Jim", 171234, "Math")
s1
```

## (2)

```python
students = [
    Student("L", "F", 1001, "Engineering"),
    Student("L", "F", 1001, "Physics"),
    Student("L", "F", 1001, "Engineering"),
    Student("L", "F", 1001, "Engineering"),
    Student("L", "F", 1001, "Physics"),
    Student("L", "F", 1001, "Math"),
]
```

```python
def count_major(stds):
    majors = {}
    for i in range(len(stds)):
        if stds[i].major in majors:
            majors[stds[i].major] += 1
        else:
            majors[stds[i].major] = 1
    return majors
```

```python
count_major(students)
```

```python
np.random.seed(0)
df1 = pd.DataFrame({
    "name": ["Sophia", "Emily", "Lily", "Olivia", "Amelia"],
    "id": range(1, 6),
    "score": np.random.randint(1, 100, 5)
})
df1
```

```python
np.random.seed(1)
df2 = pd.DataFrame({
    "name": ["Alfie", "Charlie", "Isla", "Isabella", "Ava"],
    "id": range(1, 6),
    "score": np.random.randint(1, 100, 5)
})
df2
```

```python
df1["class"] = ["A"]*5
df1
```

```python
df2["class"] = ["B"]*5
df2
```

```python
temp = pd.concat([df1, df2])
temp
```

## (1) answer

```python
ans1 = temp.set_index(["class","id","name"])
ans1
```

## (2)

```python
ans2_temp = ans1.reset_index("name", drop=True)
ans2_temp
```

```python
ans2 = ans2_temp.unstack("class")
ans2
```

## (2) answer

```python
ans2.loc["mean",:] = ans2.mean()
ans2
```
